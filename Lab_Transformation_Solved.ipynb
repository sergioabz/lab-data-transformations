```python
answer_rate = marketing_customer_analysis.copy()
answer_rate = pd.DataFrame(marketing_customer_analysis[['Response','Sales Channel',]].value_counts(dropna=False).reset_index())
answer_rate.columns = ['Response','Sales Channel', 'Rate']
answer_rate 
```

```python

answer_rate['Rate'] = (answer_rate['Rate']/marketing_customer_analysis.shape[0])*100
answer_rate
```

```python
fig, ax = plt.subplots()
ax.set_ylabel("Count")
ax.set_title("Rate of Responses by Channel %")
ax.bar(x = answer_rate['Sales Channel'], height = answer_rate['Rate'], color=['steelblue', 'grey', 'green', 'red'])

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x = answer_rate['Response'], height= answer_rate['Rate'])
ax.bar(x = answer_rate['Response'], height= answer_rate['Rate'])
ax.set_xlabel("Responses")
ax.set_ylabel("Count")
ax.set_title("total number of responses")
plt.show()
```

```python
answer_rate2 = marketing_customer_analysis.copy()
answer_rate2 = pd.DataFrame(marketing_customer_analysis[['Response','Total Claim Amount',]].value_counts(dropna=False).reset_index())
answer_rate2.columns = ['Response','Total Claim Amount', 'Rate']
answer_rate2 
```

```python
fig, ax = plt.subplots(figsize=(10,5))
# sns.barplot(x=marketing_customer_analysis['index'].value_counts().index, height = rate_df['Response'].values, hue='Sales Channel')
sns.barplot(x='Rate', y='Total Claim Amount', data=answer_rate2 )
ax.set_xlabel("Responses")
ax.set_ylabel("Count")
ax.set_title("% COUNT of responses by total claim amount")
plt.show()
```
